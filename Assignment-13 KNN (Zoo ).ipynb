# Features and Target

```python
X = df[["hair","eggs"]]
y = df["type"].values
```

```python
import matplotlib.pyplot as plt
plt.scatter(X.iloc[:,0],X.iloc[:,1])
plt.xlabel("hair")
plt.ylabel("eggs")
plt.title("Relation between hair & eggs")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(X.iloc[y==0,0],X.iloc[y==0,1],color='red',label="No type")
plt.scatter(X.iloc[y==1,0],X.iloc[y==1,1],color='green',label="type")


plt.xlabel("hair")
plt.ylabel("eggs")
plt.legend()
plt.title("Relation between hair & eggs")
plt.show()
```

# standarization

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
x=sc.fit_transform(X)
```

# Train-Test spliting

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(x,y,train_size=0.75,random_state=0)
```

# Model creation

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=2)
model.fit(X_train,y_train)
```

# Model validation : Confusion matrix

```python
y_pred = model.predict(X_test)
y_pred
```

```python
model_data = pd.DataFrame({"Actual":y_test,"Predicted":y_pred})
model_data
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

# Test- data model visualization

```python
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test


X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))

plt.contourf(X1, X2, model.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))


plt.scatter(X_set[y_set==0,0],X_set[y_set==0,1],label="No type",color='red')
plt.scatter(X_set[y_set==1,0],X_set[y_set==1,1],label="type",color='green')
plt.legend()
plt.xlabel("hair")
plt.ylabel("eggs")
plt.show()
```

# Trainning data visualization

```python
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train


X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))

plt.contourf(X1, X2, model.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))


plt.scatter(X_set[y_set==0,0],X_set[y_set==0,1],label="No type",color='red')
plt.scatter(X_set[y_set==1,0],X_set[y_set==1,1],label="type",color='green')
plt.legend()
plt.xlabel("hair")
plt.ylabel("eggs")
plt.show()
```

# Choose best number of k

```python
est = []

for i in range(1,20):
    model = KNeighborsClassifier(n_neighbors=i)
    est.append(model)
```

```python
est
```

```python
score = []

for i in range(len(est)):
    est[i].fit(X_train,y_train)
    sc = est[i].score(X_test,y_test)
    score.append(sc)
```

```python
import matplotlib.pyplot as plt
plt.scatter(range(1,20),score)
plt.plot(range(1,20),score)
plt.xlabel("Number of K")
plt.ylabel(" Accuracy")
plt.grid()
plt.show()
```
